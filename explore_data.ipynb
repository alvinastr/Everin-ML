##Importing Necessary library##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import requests
from io import StringIO
```

```python
# Download & load data from Github
url1 = "https://raw.githubusercontent.com/alvinastr/Everin/refs/heads/main/data/raw/food_data_clean.csv"
url2 = "https://raw.githubusercontent.com/alvinastr/Everin/refs/heads/main/data/raw/nutrition.csv"
```

```python
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
print("Dataset 1 Shape:", df1.shape)
print("Dataset 2 Shape:", df2.shape)
```

```python
def explore_dataset(df, name):
    print(f"\n=== Dataset: {name} ===")
    print("\nFirst few rows:")
    display(df.head())
    print("\nData Info:")
    display(df.info())
    print("\nBasic Statistics:")
    display(df.describe())
    print("\nMissing Values:")
    display(df.isnull().sum())

explore_dataset(df1, "Food Data 1")
explore_dataset(df2, "Food Data 2")
```

```python
print("\n=== Comparing Datasets ===")
print("Columns in Food Data 1:", df1.columns.tolist())
print("Columns in Food Data 2:", df2.columns.tolist())

common_columns = set(df1.columns) & set(df2.columns)
print("\nCommon Columns:", common_columns)

if 'food_id' in df1.columns and 'food_id' in df2.columns:
    print("\nUnique food_id count: ")
    print("Food Data 1:", df1['food_id'].nunique())
    print("Food Data 2:", df2['food_id'].nunique())

    common_foods = set(df1['food_id']) & set(df2['food_id'])
    print(f"\nNumber of common food_id: {len(common_foods)}")
```

```python
def plot_distributions(df1, df2, column):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(df1[column], kde=True)
    plt.title(f'{column} Distribution - Dataset 1')

    plt.subplot(1, 2, 2)
    sns.histplot(df2[column], kde=True)
    plt.title(f'{column} Distribution - Dataset 2')

    plt.tight_layout()
    plt.show()

common_numeric_columns = set(df1.select_dtypes(include=[np.number]).columns) & \
                         set(df2.select_dtypes(include=[np.number]).columns)

for column in common_numeric_columns:
    plot_distributions(df1, df2, column)
```

```python
def plot_correlation_matrix(df, title):
    numeric_columns = df.select_dtypes(include=[np.number]).columns
    corr_matrix = df[numeric_columns].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title(title)
    plt.show()

plot_correlation_matrix(df1, 'Correlation Matrix - Dataset 1')
plot_correlation_matrix(df2, 'Correlation Matrix - Dataset 2')
```

```python
if 'id' in df1.columns and 'id' in df2.columns:
    merged_df = pd.merge(df1, df2, on='_id', how='outer', suffixes=('_df1', '_df2'))
    print("\n=== Merged Dataset ===")
    print("Shape:", merged_df.shape)
    print("\nFirst few rows:")
    display(merged_df.head())
else:
    print("\nCannot merge datasets: 'food_id' not found in both datasets")
    merged_df = None
```
